```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
```

```python
#Data Collection and analysis
df=pd.read_csv(r"C:\Users\coord\OneDrive\Desktop\Mall_Customers.csv")
df.head()
```

```python
#finding theno. of rows and columns
df.shape
```

```python
df.info()
```

```python
#checking missing values
df.isnull().sum()
```

```python
label_encoder=LabelEncoder()
df['Gender']=label_encoder.fit_transform(df['Gender'])
```

```python
#feature selection (IMPORTANT)
x=df[['Annual Income (k$)', 'Spending Score (1-100)']].values
```

```python
x
```

```python
#chossing number of clusters by using WCSS Formula
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, len(wcss) + 1), wcss)
plt.title('Elbow Point Graph')
plt.xlabel('No.of cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
#optimum number of cluster are 4 
```

```python
#training of K-Means Model
kmeans=KMeans(n_clusters=4, init='k-means++',random_state=42)

#return a label for eanh data point based an their cluster
y=kmeans.fit_predict(x)
print(y)
```

```python
#visualizatin
plt.figure(figsize=(8,8))

plt.scatter(x[y==0, 0], x[y==0, 1], s=50, label='Cluster 1')
plt.scatter(x[y==1, 0], x[y==1, 1], s=50, label='Cluster 2')
plt.scatter(x[y==2, 0], x[y==2, 1], s=50, label='Cluster 3')
plt.scatter(x[y==3, 0], x[y==3, 1], s=50, label='Cluster 4')

#centroids
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    s=200, marker='X', label='Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
